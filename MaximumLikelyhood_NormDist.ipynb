```python
import numpy as np
from scipy.stats import norm
```

```python
def mle_norm(samples : list) -> int:
  probs = []
  potential_miu = np.arange(-10,10,0.01)
  for miu in potential_miu :
    probs.append(np.product(norm.pdf(samples,loc = miu)))
  # print(probs)
  return potential_miu[np.argmax(probs)]
```

```python
mle_norm ( norm.rvs(loc= 1.2 , size = 100))
```

```python
mle_norm ( norm.rvs(loc= 2.1 , size = 3000)) # --> as there are a lot of probabilities to be multiplied and they are small, python estimates them as zero
```
