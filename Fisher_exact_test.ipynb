```python
import numpy as np
from math import factorial
import matplotlib.pyplot as plt
```

```python
def comb(n , k):
  return factorial(n)//(factorial(k)*factorial(n-k))

def calc_prob_h0 (d1 : int, d2 : int,b1 : int,b2 : int):
  return comb(d1+d2 , d1) * comb(b1+b2,b1)/ comb(d1+d2+b1+b2,d1+b1)

def p_value(d1 : int, d2 : int,b1 : int,b2 : int): # Calculates p_value for fisher test
  p_val = 0
  l= []
  for i in range(min(d2,b1)+1):
    # p_val += calc_prob_h0(extreme,b1-(extreme-d1),b1,b2)
    l.append(calc_prob_h0(d1+i , d2 - i , b1 - i , b2 + i))

  # return p_val
  return l
```
