```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import seaborn as sns
```

```python
def mle_bern (samples : list):
  probs = []
  potential_ps = np.arange(0,1,0.01)
  for p in potential_ps :
    each_prob = bernoulli.pmf(samples, p) # this will return probability of each sample if p = p 
    probs.append(np.product(each_prob)) # as each item is independant, we can multiply probabilities to find chance of sequence happening
  return potential_ps[np.argmax(probs)]
```

```python
mle_bern([1,1,1,0,0,0])
```

```python
mle_bern(bernoulli.rvs(p = 0.32 , size = 100))
```

```python
# Note: mle estimator is also random variable
mle_estimated = []
for i in range(10000):
  mle_estimated.append(mle_bern(bernoulli.rvs(p = 0.34,size = 100)))
```

```python
sns.histplot(mle_estimated)
```
